Use shelve module to create Student database to store details of the students(at least 10 student details)

```python
import shelve
students = shelve.open("testdb")

while True:
    roll_no = input("Roll No of Student (Type 'exit' to quit the program): ")
    if roll_no == 'exit':
        break
    else:
        name = input("Enter Name: ")
        cgpa = float(input("Enter CGPA: "))
        dept = input("Enter Department: ")
    
    student_dict = {'Name': name, 'CGPA': cgpa, 'Department': dept}
    students[roll_no] = student_dict
    
students.close()
```

```python
dbFileName = "testdb"
shelf = shelve.open(dbFileName)

for student, data in shelf.items():
    print("Student Name: {}".format(data['Name']))
    print("Roll No: {}".format(student))
    print("CGPA: {}".format(data['CGPA']))
    print("Department: {}".format(data['Department']))
    print()
    
shelf.close()
```

```python
dbFileName = "testdb"
shelf = shelve.open(dbFileName)

def read(student):
    try:
        student_data = shelf[student].copy()
        print("Enter your choice: ")
        print("1. CGPA")
        print("2. Name")
        print("3. Department")
        print("4. All")
        
        opt = int(input("Option: "))
        print()
        
        if opt == 1:
            print("CGPA: {}".format(student_data['CGPA']))
        elif opt == 2:
            print("Name: {}".format(student_data['Name']))
        elif opt == 3:
            print("Department: {}".format(student_data['Department']))
        elif opt == 4:
            print("Name: {}".format(student_data['Name']))
            print("CGPA: {}".format(student_data['CGPA']))
            print("Department: {}".format(student_data['Department']))
        else:
            print("Invalid Option")
        
        print()
        
    except:
        print("\nRoll No Does Not Exists\n")


def update(student):
    try:
        student_data = shelf[student].copy()
        print("Enter the property you want to Update: ")
        print("1. CGPA")
        print("2. Name")
        print("3. Department")
        print("4. Roll No")
        
        opt = int(input("Option: "))
        print()
        
        if opt == 1:
            new_cgpa = float(input("Enter new CGPA: "))
            student_data['CGPA'] = new_cgpa
            shelf[student] = student_data
        elif opt == 2:
            new_name = input("Enter new Name: ")
            student_data['Name'] = new_name
            shelf[student] = student_data
        elif opt == 3:
            new_dept = input("Enter new Department: ")
            student_data['Department'] = new_dept
            shelf[student] = student_data
        elif opt == 4:
            new_roll = input("Enter new Roll: ")
            shelf[new_roll] = student_data
            del shelf[student]
        else:
            print("Invalid Option")
        
        print()
        
    except:
        print("\nRoll No Does Not Exists\n")

def delete(student):
    try:
        del shelf[student] 
    except:
        print("\nRoll No Does Not Exists\n")


while True:
    print("1. Read Data")
    print("2. Update Data")
    print("3. Delete Data")
    opt = input("Choose from 1 to 3 (Press any Other Key to exit): ")
    
    if opt == '1' or opt == '2' or opt == '3':
        student = input("Enter Roll No of student: ")
    
    if opt == '1':
        read(student)
    elif opt == '2':
        update(student)
    elif opt == '3':
        delete(student)
    else:
        break
        
shelf.close()
```
