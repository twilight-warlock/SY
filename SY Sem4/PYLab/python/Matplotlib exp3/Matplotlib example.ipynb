```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(7)
print(x)

print(plt.style.available)
plt.style.use('seaborn')

plt.plot(x,(x*x),color='blue',marker='o',label='square of values')#'bo--'
plt.plot(x,x,color='red',marker='v',linestyle='dashed',label='linear')
plt.plot(x,(x*x*x),color='green',linestyle='--',label='cube of values')

#y=x*3

#a_point = 50


#filling plots
#plt.fill_between(x,(x*x*x),y_point,alpha=0.25)

#plt.fill_between(x,(x*x*x),a_point,where=((x*x*x) > y_point),alpha=0.25, interpolate = True, color='red')

#plt.fill_between(x,(x*x*x),(x*x),where=((x*x*x) > (x*x)),alpha=0.25, interpolate = True, color='red')

plt.xlabel('number')
plt.ylabel('square')
plt.title('Example of plotting lines')

plt.legend()
plt.savefig('sampleplot.jpg')
plt.show()
```

```python
#bar charts
import matplotlib.pyplot as plt

import numpy as np
x=np.array([6,3,8,9,2,4]) 

print(x)

#print(plt.style.available)
plt.style.use('seaborn')

#plt.bar(x,(x*x),color='blue',label='square of values')
#plt.bar(x,x,color='red',label='linear')
#plt.bar(x,(x+x+x),color='green',label='cube of values')


x_indexes = np.arange(len(x))
w=0.3

plt.bar(x_indexes-w,(x*x),width=w,color='blue',label='square of values')
plt.bar(x_indexes,x,width=w,color='red',label='linear')
plt.bar(x_indexes+w,(x+x+x),width=w,color='green',label='cube of values')

plt.xticks(ticks=x_indexes,labels=x)

plt.xlabel('number')
plt.ylabel('square')
plt.title('Bar chart example')
plt.legend()

plt.show()
```

```python
#pie charts

import matplotlib.pyplot as plt

plt.style.use('seaborn')

slices = [120,30,50,80,90]
labels = ['python','javascript','C#','php','java']
colors = [ '#ff8883','#bc345d','#ef238c','#783db2','#45b34d']
explode=[0,0.2,0,0,0]

plt.pie(slices, labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},explode=explode,startangle=230,shadow=True,autopct='%1.1f%%')
#wedgeprops={'edgecolor':'black'},startangle=230,autopct='%1.1f%%',shadow=True,explode=explode

plt.title("Pie chart example")
plt.tight_layout()
plt.show()
```

```python
#stack plots /area plots
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

days = [1,2,3,4,5,6]

dev1 = [2,2,4,4,5,8]
dev2 = [1,1,1,1,2,2]
dev3 = [2,2,3,3,4,5]

labels =['developer1','developer2','developer3']

plt.stackplot(days,dev1,dev2,dev3,labels=labels)

plt.legend(loc='upper left')

plt.title("Stack Plot example")
plt.tight_layout()
plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')


marks = [35,20,47,69,50,78,69,61,50]

#plt.hist(marks,bins=4,edgecolor='black')

bins=[20,30,40,50,60,70]
plt.hist(marks,bins=bins,edgecolor='black')

#plt.legend()

plt.title("Histogram Example")
plt.xlabel('marks')
plt.ylabel('no. of students')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)#having multiple subplots within a single figure ->returns figure and axes
x = np.arange(10)

ax[0, 0].plot(x,x, 'g') #row=0, col=0
ax[0, 0].grid(True)

ax[1, 0].plot(x,(x*x), 'k') #row=1, col=0

ax[0, 1].plot(x,(x*x*x), 'b') #row=0, col=1
ax[0, 1].grid(color='g',linestyle='--',linewidth=3)


ax[1, 1].plot(x, (x*5), 'r') #row=1, col=1
fig.show()
```
